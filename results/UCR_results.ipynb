```python
import os
import numpy as np
import pandas as pd

datasets = ['ChlorineConcentration', 'PenDigits', 'PhonemeSpectra', 'Wafer']
```

```python
models = ["Ladder", "Laguerre", "LMU", "BRC", "nBRC"]

for dataset in datasets:
    max_perf = 0
    best_perf = 'None'
    
    for model in models:

        df_results = pd.read_csv(os.path.join(dataset, f'{model}_results_k-Folds.csv'))

        if dataset is not "Wafer":
            print(f"Model: {model} ({dataset})")
            print(f"Average F1: {df_results['F1'].mean():.4f}")
            print(f"Std F1: {df_results['F1'].std():.4f}\n")
            
            max_perf = max(max_perf, df_results['F1'].mean())
            if max_perf == df_results['F1'].mean():
                best_perf = model
            
        else:
            print(f"Model: {model} ({dataset})")
            print(f"Average AUROC: {df_results['AUROC'].mean():.4f}")
            print(f"Std AUROC: {df_results['AUROC'].std():.4f}\n") 
            
            max_perf = max(max_perf, df_results['AUROC'].mean())
            if max_perf == df_results['AUROC'].mean():
                best_perf = model
            
    print(best_perf, max_perf)
```
