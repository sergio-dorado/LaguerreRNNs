```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Time vector for training and testing datasets
t_training = np.linspace(0, (35/100)*2*np.pi, 750)
t_testing = np.linspace((35/100)*2*np.pi, 2*np.pi, 250)

sin_training = np.sin(t_training)
sin_testing = np.sin(t_testing)

# Plotting Laguerre Polynomials
font_plot = "Times"
fig, axes = plt.subplots(figsize = (8,6))

axes.plot(t_training, sin_training, label = "Training Set", color = 'salmon')
axes.plot(t_testing, sin_testing, label = "Testing Set", color = 'royalblue')
axes.legend(prop = {'size' : 16, 'family' : font_plot}, frameon = False)

axes.set_xlabel("Sample Index", fontname = font_plot, color = "white", fontsize = 15)
#axes.set_ylabel("Amplitude", fontname = font_plot, color = "white", fontsize = 15)
axes.set_title("Forecasting Experiment Data Split", fontname = font_plot, fontsize = 20, color = "white")

axes.tick_params(color = "white", labelcolor = "white")
plt.tight_layout()

axes.spines['bottom'].set_color('white')
axes.spines['top'].set_color('white') 
axes.spines['right'].set_color('white')
axes.spines['left'].set_color('white')

for tick in axes.xaxis.get_major_ticks():
    tick.label.set_fontsize(0)

for tick in axes.yaxis.get_major_ticks():
    tick.label.set_fontsize(0)
    
leg = axes.get_legend()
for text in leg.get_texts():
    text.set_color("white")

fig.savefig("Fig_Forecasting_Experiment_Data_Split.png", dpi = 300, transparent = True)
```
