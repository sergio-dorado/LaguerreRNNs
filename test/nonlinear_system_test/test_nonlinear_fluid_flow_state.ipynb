```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

The next cell defines the `pend` function that contains the system of nonlinear differential equations representing the pendulum dynamics.

```python
def fluid_flow(x, t, mu, omega, A, l):
    x1, x2, x3 = x 
    dxdt = [mu*x1 - omega*x1 + A*x1*x3,
           omega*x1 - omega*x2 + A*x2*x3,
           -l*(x3 - x1**2 - x2**2)]
    return dxdt
```

In the cell below, the parameters of the pendulum dynamics, the initial condition vector and the time vector are defined. 

```python
mu = 0.1
omega = 1
A = -0.1
l = 10

x0 = np.random.rand(3, )

t = np.linspace(0, 10, 1001)
```

Below, we compute the numerical solution using the `odeint` function.

```python
sol = odeint(fluid_flow, x0, t, args = (mu, omega, A, l))
```

The output is visualized extracting the states from each column of the output matrix from the integration step

```python
x1 = sol[:, 0]
x2 = sol[:, 1]
x3 = sol[:, 2]

fig, axes = plt.subplots(figsize = (10, 6))

axes.plot(t, x1, 'b', label = "$x_1(t)$", linewidth = 3)
axes.plot(t, x2, 'r', label = "$x_2(t)$", linewidth = 3)
axes.plot(t, x3, 'green', label = "$x_3(t)$", linewidth = 3)

axes.legend(loc = "best")
axes.set_xlabel("$t$");
```
